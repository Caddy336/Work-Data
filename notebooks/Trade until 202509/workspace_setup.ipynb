## 1. еҲӣе»әе·ҘдҪңз©әй—ҙзӣ®еҪ•з»“жһ„

йҰ–е…ҲпјҢжҲ‘д»¬е°ҶеҲӣе»әж ҮеҮҶзҡ„йЎ№зӣ®зӣ®еҪ•з»“жһ„пјҡ

```python
import os
from pathlib import Path

# е®ҡд№үйЎ№зӣ®ж №зӣ®еҪ•
ROOT_DIR = Path.cwd().parent

# еҲӣе»әеҝ…иҰҒзҡ„зӣ®еҪ•
directories = [
    'config',            # й…ҚзҪ®ж–Үд»¶
    'dashboards',        # ж•°жҚ®еҸҜи§ҶеҢ–йқўжқҝ
    'data/raw',         # еҺҹе§Ӣж•°жҚ®
    'data/processed',   # еӨ„зҗҶеҗҺзҡ„ж•°жҚ®
    'data/interim',     # дёӯй—ҙж•°жҚ®
    'notebooks',        # Jupyter notebooks
    'src',             # жәҗд»Јз Ғ
    'logs',            # ж—Ҙеҝ—ж–Үд»¶
    'templates',       # жЁЎжқҝж–Үд»¶
]

# еҲӣе»әзӣ®еҪ•
for dir_path in directories:
    full_path = ROOT_DIR / dir_path
    full_path.mkdir(parents=True, exist_ok=True)
    print(f"еҲӣе»әзӣ®еҪ•: {full_path}")

# йӘҢиҜҒзӣ®еҪ•з»“жһ„
print("\nйЎ№зӣ®зӣ®еҪ•з»“жһ„:")
def print_tree(directory, prefix=""):
    paths = sorted(directory.glob("*"))
    for i, path in enumerate(paths):
        if '.git' not in str(path):  # жҺ’йҷӨ.gitзӣ®еҪ•
            is_last = i == len(paths) - 1
            print(f"{prefix}{'в””в”Җв”Җ ' if is_last else 'в”ңв”Җв”Җ '}{path.name}")
            if path.is_dir():
                extension = "в””в”Җв”Җ " if is_last else "в”ңв”Җв”Җ "
                print_tree(path, prefix=prefix + ("    " if is_last else "в”Ӯ   "))

print_tree(ROOT_DIR)
```

## 2. й…ҚзҪ®зҺҜеўғеҸҳйҮҸ

и®ҫзҪ®е’ҢйӘҢиҜҒеҝ…иҰҒзҡ„зҺҜеўғеҸҳйҮҸпјҡ

```python
import os
from pathlib import Path

# е®ҡд№үйЎ№зӣ®ж №зӣ®еҪ•пјҲеҰӮжһңиҝҳжңӘе®ҡд№үпјү
if 'ROOT_DIR' not in globals():
    ROOT_DIR = Path.cwd().parent
print(f"йЎ№зӣ®ж №зӣ®еҪ•: {ROOT_DIR}")
```

```python
import os
from dotenv import load_dotenv

# еҲӣе»ә.envж–Үд»¶
env_path = ROOT_DIR / '.env'  # дҝ®жӯЈж–Үд»¶еҗҚ
if not env_path.exists():
    env_content = f"""
PROJECT_ROOT={ROOT_DIR}
DATA_DIR={ROOT_DIR}/data
LOGS_DIR={ROOT_DIR}/logs
CONFIG_DIR={ROOT_DIR}/config
"""
    env_path.write_text(env_content.strip())
    print("еҲӣе»әдәҶ.envж–Үд»¶")  # жӣҙж–°жҸҗзӨәдҝЎжҒҜ

# еҠ иҪҪзҺҜеўғеҸҳйҮҸ
load_dotenv(env_path)

# йӘҢиҜҒзҺҜеўғеҸҳйҮҸ
env_vars = ['PROJECT_ROOT', 'DATA_DIR', 'LOGS_DIR', 'CONFIG_DIR']
for var in env_vars:
    value = os.getenv(var)
    print(f"{var}: {value}")
```

```python
# е®үиЈ…python-dotenvеҢ…
import sys
import subprocess

def install_package(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

try:
    import dotenv
    print("python-dotenv е·Іе®үиЈ…")
except ImportError:
    print("жӯЈеңЁе®үиЈ… python-dotenv...")
    install_package("python-dotenv")
    print("е®үиЈ…е®ҢжҲҗ")
```

## 3. и®ҫзҪ®е·ҘзЁӢдҫқиө–йЎ№

е®үиЈ…е’ҢйӘҢиҜҒеҝ…иҰҒзҡ„PythonеҢ…пјҡ

```python
import pkg_resources
import subprocess

# е®ҡд№үеҝ…иҰҒзҡ„еҢ…
required_packages = [
    'numpy',
    'pandas',
    'matplotlib',
    'seaborn',
    'scikit-learn',
    'jupyter',
    'python-dotenv'
]

# жЈҖжҹҘе·Іе®үиЈ…зҡ„еҢ…
installed_packages = {pkg.key: pkg.version for pkg in pkg_resources.working_set}

# жЈҖжҹҘе’Ңе®үиЈ…зјәеӨұзҡ„еҢ…
missing_packages = []
for package in required_packages:
    if package not in installed_packages:
        missing_packages.append(package)
        print(f"зјәеӨұеҢ…: {package}")

if missing_packages:
    print("\nе®үиЈ…зјәеӨұзҡ„еҢ…...")
    subprocess.check_call(['pip', 'install'] + missing_packages)
else:
    print("\nжүҖжңүеҝ…иҰҒзҡ„еҢ…йғҪе·Іе®үиЈ…")

# жҳҫзӨәе·Іе®үиЈ…еҢ…зҡ„зүҲжң¬
print("\nе·Іе®үиЈ…зҡ„еҢ…зүҲжң¬пјҡ")
for package in required_packages:
    version = installed_packages.get(package)
    print(f"{package}: {version}")
```

## 4. еҲқе§ӢеҢ–ж•°жҚ®еӯҳеӮЁдҪҚзҪ®

еҲӣе»әе№¶йӘҢиҜҒж•°жҚ®еӯҳеӮЁзӣ®еҪ•пјҡ

```python
import json

# ж•°жҚ®зӣ®еҪ•й…ҚзҪ®
data_config = {
    'raw_data_dir': str(ROOT_DIR / 'data/raw'),
    'processed_data_dir': str(ROOT_DIR / 'data/processed'),
    'interim_data_dir': str(ROOT_DIR / 'data/interim')
}

# еҲӣе»әж•°жҚ®й…ҚзҪ®ж–Үд»¶
config_dir = ROOT_DIR / 'config'
config_file = config_dir / 'data_config.json'

config_dir.mkdir(exist_ok=True)
with open(config_file, 'w') as f:
    json.dump(data_config, f, indent=4)

print("ж•°жҚ®зӣ®еҪ•й…ҚзҪ®е·ІдҝқеӯҳеҲ°:", config_file)

# йӘҢиҜҒж•°жҚ®зӣ®еҪ•жҳҜеҗҰеҸҜи®ҝй—®
for dir_name, dir_path in data_config.items():
    path = Path(dir_path)
    if path.exists() and path.is_dir():
        print(f"вң“ {dir_name}: {dir_path} (еҸҜи®ҝй—®)")
    else:
        print(f"вң— {dir_name}: {dir_path} (дёҚеҸҜи®ҝй—®)")
```

## 5. и®ҫзҪ®ж—Ҙеҝ—и®°еҪ•

й…ҚзҪ®loggingжЁЎеқ—пјҡ

```python
import logging
import logging.config
import yaml
from datetime import datetime

# еҲӣе»әж—Ҙеҝ—й…ҚзҪ®
log_config = {
    'version': 1,
    'disable_existing_loggers': False,
    'formatters': {
        'standard': {
            'format': '%(asctime)s [%(levelname)s] %(name)s: %(message)s'
        },
    },
    'handlers': {
        'file': {
            'class': 'logging.FileHandler',
            'filename': str(ROOT_DIR / 'logs' / f'app_{datetime.now().strftime("%Y%m%d")}.log'),
            'formatter': 'standard',
            'level': 'DEBUG',
        },
        'console': {
            'class': 'logging.StreamHandler',
            'formatter': 'standard',
            'level': 'INFO',
        }
    },
    'root': {
        'handlers': ['console', 'file'],
        'level': 'DEBUG',
    }
}

# дҝқеӯҳж—Ҙеҝ—й…ҚзҪ®
config_path = ROOT_DIR / 'config' / 'logging.yaml'
with open(config_path, 'w') as f:
    yaml.dump(log_config, f)

# й…ҚзҪ®ж—Ҙеҝ—зі»з»ҹ
logging.config.dictConfig(log_config)
logger = logging.getLogger(__name__)

# жөӢиҜ•ж—Ҙеҝ—
logger.debug('иҝҷжҳҜдёҖжқЎи°ғиҜ•ж—Ҙеҝ—')
logger.info('иҝҷжҳҜдёҖжқЎдҝЎжҒҜж—Ҙеҝ—')
logger.warning('иҝҷжҳҜдёҖжқЎиӯҰе‘Ҡж—Ҙеҝ—')

print(f"\nж—Ҙеҝ—й…ҚзҪ®е·ІдҝқеӯҳеҲ°: {config_path}")
print(f"ж—Ҙеҝ—ж–Үд»¶е°ҶдҝқеӯҳеңЁ: {ROOT_DIR}/logs/")
```
